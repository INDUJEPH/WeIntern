# Task 1: Sales Data Analysis Dashboard 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sampleSuperstore.csv")
```

```python
print("Dataset Info :")
df.info()
```

```python
print("Summary Statistics :")
df.describe()
```

```python
print("Missing Values :")
df.isnull().sum()
```

##### dropping rows where sales or profit is missing

#### Visualization 1: sales by region

```python
plt.figure()
sales_by_region = df.groupby('Region')['Sales'].sum()
sales_by_region.plot(kind='pie' , autopct='%1.1f%%')
plt.title('total sales by region')
plt.tight_layout()
plt.show()
```

#### Observation:
##### The West region records the highest total sales, showing strong customer demand and better market performance.
##### The South region has the lowest sales, which indicates potential opportunities for growth through improved marketing strategies.

#### Visualization 2: sales by category

```python
plt.figure()
sales_by_region = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
sales_by_region.plot(kind='bar')
plt.title('total sales by category')
plt.xlabel('category')
plt.ylabel('total sales')
plt.tight_layout()
plt.show()
```

#### Observation:
##### Technology is the highest-selling category, followed by Furniture.
##### Office Supplies contribute comparatively lower sales, but they may generate steady revenue through frequent purchases.

#### Visualization 3: top 10 categories 

```python
plt.figure()
top_subcategories = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
top_subcategories.plot(kind='barh')
plt.title('top 10 sub categories ')
plt.xlabel('Sales')
plt.ylabel('sub_categories')
plt.tight_layout()
plt.show()
```

#### Observation:
##### Phones and Chairs are the top-performing sub-categories, contributing the most to overall sales.
##### Other sub-categories have significantly lower sales, showing uneven demand across products.

#### Visualization 4: discount vs profit 


```python
plt.figure()
plt.scatter(df['Discount'], df['Profit'])
plt.title('discount vs profit')
plt.xlabel('discount')
plt.ylabel('profit')
plt.tight_layout()
plt.show()
```

#### Observation:
##### There is a clear negative relationship between discount and profit.
##### Higher discounts often result in reduced profit or losses, indicating that excessive discounting affects business profitability.

#### Visualization 5 : sales distribution by category

```python
plt.figure()
sns.boxplot(x = 'Category' , y ='Sales', data =df)
plt.title('category')
plt.xlabel('category')
plt.ylabel('sales')
plt.tight_layout()
plt.show()
```

#### Observation:
##### Technology shows a wider sales distribution, meaning it includes both high and low sales values.
##### Office Supplies have a more consistent and stable sales distribution compared to other categories.

#### Overall Conclusion:

##### West region is the best-performing region

##### Technology category generates maximum sales

##### High discounts reduce profit

##### Business shows stable growth patterns
